### Description

For this project, we will use data from two separate files. The dataset `players.csv` consists of $9$ variables and $196$ observations. This dataset contains a list of every registered player along with the data they provided when registering, with each observation being a unique player. The second dataset `sessions.csv` consists of $5$ variables and $1535$ observations. Each observation is a play session of a given player.

### Players Dataset

The $9$ variables of the Players dataset consist of the following:
- `experience` (chr type): Self-declared skill level of the player - one of `Amateur`, `Beginner`, `Regular`, `Veteran`, or `Pro` (with `Amateur` being the least experienced, and `Pro` being the most experienced)
- `subscribe` (lgl type): Indicates whether the player is subscribed to the game or not - one of `True` or `False`
- `hashedEmail` (chr type): A hashed encryption of the player's email - it can be unhashed using a key
- `played_hours` (dbl type): Total hours played by the player (in **hours**)
- `name` (chr type): Name of the player
- `gender` (chr type): Gender of the player
- `age` (dbl type): Age of the player (in **years**)
- `individialID` (lgl type): ID of the individual player (unavailable/**NA**)
- `organizationName` (lgl type): Organization of the player (unavailable/**NA**)

### Sessions Dataset

The $5$ variables of the Sessions dataset consist of the following:
- `hashedEmail` (chr type): A hashed encryption of the player's email - it can be unhashed using a key
- `start_time` (chr type): The data and time of the start of the session (**DD**/**MM**/**YYYY** and **time** format)
- `end_time` (chr type): The date and time of the end of the session (**DD**/**MM**/**YYYY** and **time** format)
- `original_start_time` (dbl type): UNIX time at the start of the session (**seconds** from 00:00:00 UTC on Thursday, 1 January 1970)
- `original_end_time` (dbl type): UNIX time at the end of the session (**seconds** from 00:00:00 UTC on Thursday, 1 January 1970)

## Question

**We would like to know which "kinds" of players are most likely to contribute a large amount of data so that we can target those players in our recruiting efforts.** The more play time a player has, the more data they contribute. Thus, `played_hours` can be a good measure of data contribution. One way to go about this question is to group all players by experience and add up their `played_hours`. Alternatively, we can also group players by their subscription status. However, it wouldn't be as viable towards the main objective of recruiting new players, as registered players who unsubscribed probably want to be left alone.

## Exploratory Data Analysis and Visualization

For this example, we're only using a couple of variables from the player dataset. We will make a graph comparing `experience` to `played_hours`, `subscribe` to `played_hours`, `gender` to `played_hours`, and `age` to `played_hours`. For the group project, we may decide to use a different approach, but for simplicity (just for a preview of what we can do with this data), we will omit the other unused variables.

```r
# Select Columns
players_select <- players |>
    select(played_hours, experience, gender, age, subscribe)

players_select
```

```r
# Group by experience
players_experience <- players_select |>
    group_by(experience) |>
    summarize(play_time = sum(played_hours))

players_experience

# Graph Played Hours versus Experience
options(repr.plot.width = 12, repr.plot.height = 6)
experience_plot <- players_experience |>
    ggplot(aes(x = experience, y = play_time)) +
    geom_bar(stat = "identity") +
    labs(title = "Play Time vs Experience", x = "Experience", y = "Play Time (hrs)") +
    theme(text = element_text(size = 20))

experience_plot
```

The Play Time versus Experience visualizations demonstrates that new players and regulars contribute the most data. With that in mind, it appears that this graph would not be very useful as it's hard to tell if a new prospective player is new to the game without them signing up.

```r
# Group by subscribe
players_subscribe <- players_select |>
    group_by(subscribe) |>
    summarize(play_time = sum(played_hours))

players_subscribe

# Graph Played Hours versus Subscribed
options(repr.plot.width = 6, repr.plot.height = 6)
subscribe_plot <- players_subscribe |>
    ggplot(aes(x = subscribe, y = play_time)) +
    geom_bar(stat = "identity") +
    labs(title = "Play Time vs Subscribed", x = "Subscribed", y = "Play Time (hrs)") +
    theme(text = element_text(size = 20))

subscribe_plot
```

As suspected, subscribed players are more likely to contribute than non-subscribed players. However, there's little control over this variable when recruiting new players.

```r
# Group by gender
players_gender <- players_select |>
    group_by(gender) |>
    summarize(play_time = sum(played_hours))

players_gender

# Graph Played Hours versus Gender
options(repr.plot.width = 12, repr.plot.height = 6)
gender_plot <- players_gender |>
    ggplot(aes(x = gender, y = play_time)) +
    geom_bar(stat = "identity") +
    labs(title = "Play Time vs Gender", x = "Gender", y = "Play Time (hrs)") +
    theme(text = element_text(size = 20))

gender_plot
```

With this visualization, we're getting somewhere. It appears that male players contribute more than any other gender. That said, this graph isn't very accurate. I believe a better measure would be to count the total number of male players, and then divide it by the total play time of male players to get a better representation. For example, what if one male player contributed $400$ hours? That would make this visualization moot.

```r
# Group by age
players_age <- players_select |>
    group_by(age) |>
    summarize(play_time = sum(played_hours))

players_age

# Graph Played Hours versus Gender
options(repr.plot.width = 12, repr.plot.height = 6)
age_plot <- players_age |>
    ggplot(aes(x = age, y = play_time)) +
    geom_point() +
    labs(title = "Play Time vs Age", x = "Age (yrs)", y = "Play Time (hrs)") +
    theme(text = element_text(size = 20))

age_plot
```

Once again, although this may not be the best visualization, age gives us a better answer to our question than experience and subscription status.
